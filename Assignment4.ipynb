```python
import cv2
print(cv2.__version__)
```

```python
#Loading a image
img=cv2.imread("c:/Users/ABC/Downloads/imageprocess.webp")
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resize a image
#img.shape
resized_img=cv2.resize(img,(500,500))
cv2.imshow('Resized Image',resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Convert image to grayscale
gray_img=cv2.cvtColor(resized_img,cv2.COLOR_BGR2GRAY)
cv2.imshow('Gray Image',gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Basic image Transformation
cropped_img=resized_img[100:200,100:400]
cv2.imshow('Crop Image',cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Rotating an Image
(h,w)=resized_img.shape[:2]
center=(w//2,h//2)
M=cv2.getRotationMatrix2D(center,37,1.0)
rotated_img=cv2.warpAffine(resized_img,M,(w,h))
cv2.imshow('Rotated Image',rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Image blurring techniques
#Gaussian Blurring

blur_img=cv2.GaussianBlur(resized_img,(5,5),0)
cv2.imshow('Gaussian Blur',blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Median Blurring

blured_img=cv2.medianBlur(resized_img,5)
cv2.imshow('MedianBlur',blured_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Bilateral Filter
blured_img=cv2.bilateralFilter(resized_img,9,75,75)
cv2.imshow('BilateralFilter',blured_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Average Blurring
blured_img=cv2.blur(resized_img,(5,5))
cv2.imshow('Average Blurring',blured_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Image Threshholding Techniques
# Binary Threshholding
_, binary_thresh=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY)
cv2.imshow('Binary Thresholding',binary_thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Inverse Binary Threshholding
_, binary_thresh=cv2.threshold(gray_img,127,255,cv2.THRESH_BINARY_INV)
cv2.imshow('Binary Thresholding',binary_thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# Truncating Threshholding
_, binary_thresh=cv2.threshold(gray_img,127,255,cv2.THRESH_TRUNC)
cv2.imshow('Binary Thresholding',binary_thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Zero Threshholding
_, binary_thresh=cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO)
cv2.imshow('Binary Thresholding',binary_thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Inverse Zero Threshholding
_, binary_thresh=cv2.threshold(gray_img,127,255,cv2.THRESH_TOZERO_INV)
cv2.imshow('Binary Thresholding',binary_thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# adaptive thresholding
adaptive_thresh_img = cv2.adaptiveThreshold(gray_img, 225, cv2.ADAPTIVE_THRESH_GAUSSIAN_C , cv2.THRESH_BINARY, 11, 2)
cv2.imshow('adsptive thresholding image', adaptive_thresh_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Edge Detection
# Canny Edge Detection
edges=cv2.Canny(resized_img,100,200)
cv2.imshow('Canny Edges',edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
