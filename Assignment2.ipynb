```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Housing.csv')

# Scatter Plot: Price vs. Area
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='area', y='price', hue='furnishingstatus', style='mainroad')
plt.title('Price vs. Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.legend()
plt.show()

```

```python
# Bubble Chart: Bedrooms vs. Bathrooms
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='bedrooms', y='bathrooms', size='price', hue='prefarea', sizes=(100, 1000), alpha=0.6)
plt.title('Bedrooms vs. Bathrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Number of Bathrooms')
plt.legend()
plt.show()
```

```python
# Pie Chart: Furnishing Status Distribution
furnishing_counts = df['furnishingstatus'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(furnishing_counts, labels=furnishing_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Furnishing Status Distribution')
plt.show()

```

```python
# Bar Chart: Average Price per Location
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='prefarea', y='price', estimator='mean', ci=None)
plt.title('Average Price per Location')
plt.xlabel('Preferred Area')
plt.ylabel('Average Price ($)')
plt.show()
```

```python
# Histogram: Distribution of Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=10, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()



```
